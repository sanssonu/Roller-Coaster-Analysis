### Identifying Duplicated Columns and Rows

```python
df.loc[df.duplicated()]
```

```python
df.loc[df.duplicated(subset=['Coaster_Name'])].head(5)
```

This shows us a list of Coaster_Names appearing for the second time. Let's see if this is true through one example.

```python
# Checking an example duplicate.
df.query('Coaster_Name == "Crystal Beach Cyclone"')
```

Yes, this is indeed true. Crystal Beach Cyclone does have two rows. All the values are same except for Year_Introduced. Maybe the roller coaster was taken down for a year and then put back up. We just want the first time a rolle coaster was introduced.

```python
# Selecting only the non-duplicate values of the specified subset for our dataframe.
df = df.loc[~df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date'])] \
       .reset_index(drop = True).copy()  # dropping rows
```

```python
df.shape
```

Now we have 990 unique roller coasters with 13 different features we're going to explore.

# 3 - Feature Understanding

(Univariate Analysis) <br>
- Plotting Feature Distributions <br>
&#9656; Histogram <br>
&#9656; KDE <br>
&#9656; Boxplot <br>

```python
df['Year_Introduced'].value_counts()
```

```python
# Bar plot for Top 10 Years Coasters Introduced.
ax = df['Year_Introduced'].value_counts() \
    .head(10) \
    .plot(kind='bar', title='Top 10 Years Coasters Introduced')
ax.set_xlabel('Year_Introduced')
ax.set_ylabel('Count')
```

&#9656; Inference: Maximum number of roller coasters were made in 1999 and 2000.

```python
# Histogram for Coaster Speed (mph).
ax = df['Speed_mph'].plot(kind='hist', 
                          bins=20, 
                          title='Coaster Speed (mph)')
ax.set_xlabel('Speed(mph)')
```

&#9656; Inference: Most of the roller coasters have a speed of 40-60 mph.

```python
# KDE plot for Coaster Speed (mph).
ax = df['Speed_mph'].plot(kind='kde', 
                          title='Coaster Speed (mph)')
ax.set_xlabel('Speed(mph)')
```

&#9656; Inference: Roller coasters hitch up at 35 mph and the majority has the highest speed of 50 mph.

# 4 - Feature Relationships

```python
# Matplotlib Scatter plot for Coaster Speed VS Height.
df.plot(kind='scatter', 
        x='Speed_mph',
        y='Height_ft',
        title='Coaster Speed VS Height')
plt.show()
```

&#9656; Inference: Speed and Height have a pretty much linear relationship, i.e, as the height of the coaster increases, its speed also increases (in most of the cases).

```python
# Seaborn Scatter plot for Coaster Speed VS Height.
sns.scatterplot(x='Speed_mph',
                y='Height_ft',
                hue='Year_Introduced',
                data=df)
plt.title('Coaster Speed VS Height')
plt.show()
```

&#9656; Inference: Same as above plot but here we also see that majority of the roller coasters were made after 1975.

```python
# Pair plots of all numeric features.
sns.pairplot(df, 
             vars=['Year_Introduced', 'Speed_mph',
                    'Height_ft', 'Inversions', 'Gforce'],
             hue='Type_Main')
plt.show()
```

```python
df_corr = df[['Year_Introduced', 'Speed_mph', 
              'Height_ft', 'Inversions', 'Gforce']].dropna().corr()
df_corr
```

This shows us the correlation between different numeric features. <br>
For eg: Speed and Height have a correlation of 73%.

```python
# Heatmap of above correlation df.
sns.heatmap(df_corr, annot=True)
```

# 5 - What are the locations with the fastest roller coasters (minimum of 10)?

```python
df['Location'].value_counts()
```

```python
# Filtering out Locations != Other
ax = df.query('Location != "Other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean', 'count']) \
    .query('count >= 10') \
    .sort_values('mean')['mean'] \
    .plot(kind='barh', figsize=(12, 5), title='Average Coaster Speed by Location')
ax.set_xlabel('Average Coaster Speed')
```

Inference: Busch Garden Williamsburg has the highest average coaster speed with more than 10 roller coasters.
